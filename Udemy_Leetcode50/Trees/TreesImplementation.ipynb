```python
class Node:
  def __init__(self,value):
    self.left = None
    self.right = None
    self.value = value

#Inorder Traversal in Left - Node - Right
def inorder(node):
  if node is not None: #If node is None it will break out of the function andd the next stepss will be executed
  #It will break and we will be pointing to the last node that called this function
    #Traverse all the way to the left side
    inorder(node.left)
    #After we break out of the fucntion at the if condition, print the current value
    print(node.value)
    #Traverse all the way to the right
    inorder(node.right)

#Preorder Traversal in Node - Left - Right
def preorder(node):
  if node is not None: #If node is None it will break out of the function andd the next stepss will be executed
  #It will break and we will be pointing to the last node that called this function
    #After we break out of the fucntion at the if condition, print the current value
    print(node.value)
    #Traverse all the way to the left side
    preorder(node.left)
    #Traverse all the way to the right
    preorder(node.right)

#Post Traversal in Left - Right - Node
def postorder(node):
  if node is not None: #If node is None it will break out of the function andd the next stepss will be executed
  #It will break and we will be pointing to the last node that called this function
    #Traverse all the way to the left side
    postorder(node.left)
    #Traverse all the way to the right
    postorder(node.right)
    #After we break out of the fucntion at the if condition, print the current value
    print(node.value)
    
    

root = Node(4)
node1 = Node(5)
node2 = Node(6)
root.left = node1
root.right = node2

node3 = Node(7)
node1.left = node3

print("--------In")
inorder(root)
print("--------Pre")
preorder(root)
print("--------Post")
postorder(root)


print("****************")

root1 = Node(10)
root1.left = Node(20)
root1.right = Node(12)
root1.left.left = Node(3)
root1.left.right = Node(8)

print("--------In")
inorder(root1)
print("--------Pre")
preorder(root1)
print("--------Post")
postorder(root1)



```

```python
"""
    BINARY SEARCH TREE:

            5
          /   \
        3.     7
       / \.   / \
      2   4. 6.  8


"""
class Node:
  def __init__(self,value):
    self.left = None
    self.right = None
    self.value = value

def insert(root,node):
  #As a good habit, check if the tree is valid / root iis valid. This will also be the break condition of the recursion loop
  if root is None:
    root = node
    return #Will end the function

  #If value of current node is less than root, go left
  if root.value < node.value :
    # We know we have to go left. So check if the current root actually has a left subtree or not
    if root.right is None:
      root.right = node
    else:
      # if a left subtree does exist, call insert function again to check conditions again
      insert(root.right,node)
  
  #Now if the node is to be aadded to the right subtree
  else:
    #Check if right subtree exists
    if root.left is None:
      root.left = node
    else:
      insert(root.left,node)

def search(node,e):
  
  if node.value == e:
    # return True
    print("True")
    return
  
  if e < node.value:
    if node.left is None:
      # return False
      print("False")
      return
    else:
      print(node.value)
      search(node.left,e)

  if e > node.value:
    if node.right is None:
      # return False
      print("False")
      return
    else:
       search(node.right,e)


#Pre order traversal
def preorder(node):
  if node is not None:
    print(node.value)
    preorder(node.left)
    preorder(node.right)

def mainFunction(list1):
  #list contains the node values randomly
  tree = Node(list1[0])

  for i in range(1,len(list1)):
    insert(tree,Node(list1[i]))
    # preorder(tree)
    # print("****")

  #preorder(tree)
  search(tree,9)

mainFunction([5,3,2,4,7,6,8])
```
