```
"""
Adding node in front and after a given node does O(1) work because the amount of work is comstant. But NodeAtEnd does O(n) work because of the loop.
This can be made O(1) by storing the pointer to last node in a variable and passing that.
"""
class Node:

  #Constructor ; so initialize node. Pointer empty
  def __init__(self,data):
    #print(data)
    self.data=data; #Means 1 data varaible hai is class mei. Global for the class
    self.next=None


class LinkedList:
  def __init__(self):
    self.head=None #Means 1 data varaible hai is class mei. Global for the class

  def printNodeVal(self):
    print(self.head)
    temp=self.head
    while(temp): 
      print(temp.data) 
      temp = temp.next

  #----ADD NODE AFTER A GIVEN NODE
  # prev_node=previous node ie node after which new_data is to be inserted
  def addNodeAfter(self, prev_node,new_data):
    if prev_node==None:
      print("The given previous node must inLinkedList")
      return
    
    else:
      new_node=Node(new_data)
      new_node.next=prev_node.next
      prev_node.next=new_node

  #----ADD NODE AT THE END
  def addAtEnd(self,new_data):
    new_node=Node(new_data)

    #If linked list doesnt exist, new Node becomes linked list
    if self.head==None:
      self.head=new_node
      new_node.next=None
    
    else:
      #Traverse
      linkList=self.head
      while(linkList.next):
        linkList=linkList.next
      linkList.next=new_node
      new_node.next=None

  def delNode(self,key):
    linkList=self.head
    #if node to be deleted is the first node

    if(linkList is not None):
      if(linkList.data==key):
        self.head=linkList.next
        linkList=None

    while(linkList is not None):
      if(linkList.data==key):
        #DELETE
        prev_node.next=linkList.next
      prev_node=linkList
      linkList=linkList.next


if __name__=='__main__': 

  llist=LinkedList() # sets head as None
  print(llist)

  #Pehle sirf create nodes. Har node k next ko none. Phir baad mei connect karo
  llist.head=Node(1) #Node1 sets data as 1 and llist.head as 1
  #print(llist.head) #Head (1st node is called head) of Linked list -> Pointer to the first node of LL. Head ki value ab 1 ho gayi
  second=Node(6) # second variable mei -> data ki value 2 set hui jab object initialize kara. Wo data ki value in second now
  third=Node(3)

  ''' 
    Three nodes have been created. 
    We have references to these three blocks as head, 
    second and third 
  
    llist.head        second              third 
         |                |                  | 
         |                |                  | 
    +----+------+     +----+------+     +----+------+ 
    | 1  | None |     | 2  | None |     |  3 | None | 
    +----+------+     +----+------+     +----+------+ 
    '''
  llist.head.next = second;
  second.next=third
  third.next=None
  #---TRAVERSAL
  #llist.printNodeVal()
  ''' 
    Now next of first Node refers to second.  So they 
    both are linked. 
  
    llist.head        second              third 
         |                |                  | 
         |                |                  | 
    +----+------+     +----+------+     +----+------+ 
    | 1  |  o-------->| 2  | null |     |  3 | null | 
    +----+------+     +----+------+     +----+------+  

    Now next of second Node refers to third.  So all three 
    nodes are linked. 
  
    llist.head        second              third 
         |                |                  | 
         |                |                  | 
    +----+------+     +----+------+     +----+------+ 
    | 1  |  o-------->| 2  |  o-------->|  3 | null | 
    +----+------+     +----+------+     +----+------+  
    '''
  #-----ADD A NODE TO THE FRONT
  front=Node(10)
  first=llist.head
  llist.head=front
  front.next=first
  first.next=second

  #llist.printNodeVal()
  llist.addNodeAfter(second,12)
  llist.printNodeVal()

  llist.addAtEnd(15)
  #llist.printNodeVal()

  llist.delNode(10)
  llist.printNodeVal()
  
```
