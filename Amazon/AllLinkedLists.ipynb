```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    #This sollutiion iis the recursive solution. 
    """
        a sub LL is essentiallly a LL. If we have a reverse function which takes in the sub LL
        and processes it like a normal llll and return the reversed ll. Once returnedd from this function,
        we will have to maake the right connection with the previous node. 
        
        1. First we have to check if there are k nnodes availlable in thhe LL. 
        2. If yes -> send the whole LL with the value of k to the reverse LL function.
            Inside the reverseLL function loop until 3 spaces and revrse the LL until k position. Send back the head. This heaad            will be stored in the reverseHead variable.
    """
    def reverseKGroup(self, head: Optional[ListNode], k: int) -> Optional[ListNode]:
        cur = head
        counter = 0
        
        #1. Count how many nodes remaining in the LL. Then if count = k that means enough available -> Make the recursive call
        #Also with this loop we are moving thhe cursor by k spots.
        while (counter < k and cur):
            cur = cur.next
            counter +=1
            
        if counter == k:
            # 2. Pass the whole head of the LL to thhe funnction. 
            reversehead = self.reverseLL(head,k)
            #Then we set the pointer of the current head value to the value returned from the recursive reverseKgroup function.
            #This function returns the reverseheadd value of the k depending on the value of k
            #Making the right connection here
            head.next = self.reverseKGroup(cur,k)
            return reversehead
            
        return head

    def reverseLL(self,head:Optional[ListNode], k: int)-> Optional[ListNode]:
        previous = None
        cur = head
        
        while (k > 0):
            nextNode = cur.next
            cur.next = previous
            previous = cur
            cur = nextNode
            k -=1
            
        return previous
```

```python
#Deep Copy of linkedd List Recursive approach
"""
    We have to copy a list aas it is - next pointing to next and raandom pointing to random
    
    Essentiallly what we are doing is as follows:
    1. We are creaating a newNoded from the value of the currentHead
    2. Then we are adding this newNode to the dictionary with head as key and the newNode as value.
        This basically means that this node has been created and no loops should be created incase a random
        pointer points backwards to a traaversed node
    3. Then we recursively call this function for the node held in the the cur.next andd cur.random.
    4. In this function, for this recursion to work, we will have to add aa check to see if the node in question
        has already been created. If yes, simply return that node from the hashmap. If it's not in thhe hashmap, the newly created          node gets returned after been added to the hashmap
        
"""
"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""
from collections import defaultdict
class Solution:
    
    #The hashmap has to be a global haashmap. Therefore:
    def __init__(self):
        self.hashmap = {}
    
    
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        #REcursive approach
        #visited_dictionary[current_node] = cloned_node_for_current_node.
        
        if head == None:
            return None
        
        if head in self.hashmap:
            return self.hashmap[head]
        
        #Create the newNode with the head value
        newNode = Node(head.val,None,None)
        #Add this newNoded as visited in the hashmap
        self.hashmap[head] = newNode
        #Copy the next and random pointers recursively
        newNode.next = self.copyRandomList(head.next)
        newNode.random = self.copyRandomList(head.random)
        #Finally return this newNode
        return newNode
        
        
        
        
        
```

```python
# Deep list Iterative approach best solution
"""
    We can improvee the previous recursive method by using an ierative approach andd without using a hashmap.
    Ssince we are not using the hashmap and not creating another linked list, our space complexity becomes O(1)
    
    Steps:
    1. Create a newNode with the cur pointer. Put this newnode next to the orginal node
        which means newNodde.next will contain ptr.next and ptr.next will point to newNode. Then move the pointer to the 
        nextNode of the original list.
        
    2. Once we have the merged llist, traverse it again and assign the random pointer of nodes from the original list to
        the next node which will be a cloned node. if no random pointer assign None. Increment the pointer by 2 spaces.
    
    3. Unweave the list to seperate original and cllonedd lists
        
"""
"""
# Definition for a Node.
class Node:
    def __init__(self, x: int, next: 'Node' = None, random: 'Node' = None):
        self.val = int(x)
        self.next = next
        self.random = random
"""
from collections import defaultdict
class Solution:
    
    
    def copyRandomList(self, head: 'Optional[Node]') -> 'Optional[Node]':
        #Iterative approach (best solution)
        if not head:
            return head
        
        ptr = head
        #-----Step 1
        while ptr:
            #Create new Cloned Node
            newNode = Node(ptr.val,None,None)
            # newNode points to original heaad's next
            newNode.next = ptr.next
            #Ptr.next points to newNode
            ptr.next = newNode
            #Increment ptr to point to next original Node
            ptr = newNode.next
            
        #point ptr to head again
        ptr = head
        
        #-----Step 2 : Assign ranndom pointers
        while ptr:
            if ptr.random:
                #Random pointer of cloned node becomes current node's random
                ptr.next.random = ptr.random.next
            else:
                ptr.next.random = None
            ptr = ptr.next.next
                
        
        #----Step 3: Unweave the linked List
        originalListPtr = head #Pointer to original list
        #ClonedLList head is the second value after head
        newHead = head.next
        cloneListPtr = newHead #Pointer to head of clonedList
         
        
        while originalListPtr:
            originalListPtr.next = originalListPtr.next.next #Original lists next becomes one node after next node
            cloneListPtr.next = cloneListPtr.next.next if cloneListPtr.next else None
            
            #Increment originalPointer
            originalListPtr = originalListPtr.next
            #Increment new Pointer
            cloneListPtr = cloneListPtr.next
            
        #Finallly return newHead
        return newHead
        
        
                
            
        
        
        
        
        
        
```

```python
"""
Given the head of a linked list, remove the nth node from the end of the list and return its head.

Step 1: We use the 2 pointer technique. First we create a node which we place BEFORE the head of the list. This node is a null node. Then we create firstpointer which we will place at
        n+1 position from the beginning of the LL. Eg: If n=2 ; firstPointer = 3 therefore, 3rd element from the beginning.
        If n=1 ; firstPointer will point to 2 ie second element from beginning
Step 2: We keep the seconddPointer aat the beginning of the llist which is the null node we created
        eg: if LL is Null -> 1 -> 2 -> 3 -> 4 -> 5 andd n=2
            firstPointer will be at node 3 and first pointer will be at Null

            Therefore,  these 2 pointers arrre always seperated by n nodes ----- IMPORTANT
Step 3: traverse the LL now at the same speed so that the difference is always maintained. We will traverse the link until the firstPointer.next becomes nulll

IMPORTANT --- We make thhe null head and pllace the secondPointer there, because when the traversing ends, this pointer willl point to one node before the node to be deleted.
              And thats the node we need whose next has to be changed for the deletetion of nth node from the back

"""
def removeNthFromEnd(self, head: Optional[ListNode], n: int) -> Optional[ListNode]:
  dummyNode = Node(0)
  dummyNode.next = head
  counter = n + 1

  firstPointer = dummyNode
  while(counter > 0):
    firstPointer = firstPointer.next
    counter -= 1

  secondPointer = dummyNode

  while firstPointer is not None:
    firstPointer = firstPointer.next
    secondpointer = secondPointer.next

  secondpointer.next = secondPointer.next.next
  return dummyNode.next
  
```
