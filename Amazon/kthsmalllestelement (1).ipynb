```python
"""
  Leetcode. 230 : kth smallest element in a BST
  Given the root of a binary search tree, and an integer k, return the kth smallest value (1-indexed) of all the values of the nodes in the tree.

  ----------Basic Inorder Traversal. is aa type of traversaal in an orddedred fashion
  Since BSTs are by defaaullt sorted ie for every node, left subtree has node less than the node aand right is greater than the node


  InorderTraversal is Left -> node -> right
  eg:         5
          3       7
        1   4
      will be 1 -> 3 -> 4 -> 5 -> 7
      This gives us ascending order of nodes. Now simply return the kth smallest ellement
    
  PseudoCode: Inorder(node,k):
      #Base case 1
      if node is invalid:
        return #With just return we go back to where the execution was left off
      
      inorder(node.left,k)

      -- So after this traversal, we have traversed to the end of the path for left subtree which is 1 in our case. ie the smallest element.
          Now we can start decrementing k

      k -= 1
      if k ==0: #ie we have reached our node
        ans = node.val
        return 

      #---some logic here
      inorder(node.right,k) #At this point k has decremented si value of k will be ddifferent

      -----------------------------------
      Time Compllexity - O(N) - searching every node
      Space Compllexity - O(N) - recursion

"""

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def kthSmallest(self, root: Optional[TreeNode], k: int) -> int:
        #Begin with creating cllass ellemennts
        self.k = k
        self.ans = None
        
        #call the inorder function with the root
        self.inorder(root)
        return self.ans
        
    def inorder(self, node):
        #Base condition 1
        if node is None:
            return
        
        #Else start inordedr traversal - start with left node.
        #When travelled till the last node on the left subtree the base condition 1 willl trigger and we will return to where
        #we left the execution
        self.inorder(node.left)
        
        #After we are returned from the left subtree, we decrement k siince we have reached the first element of the BST
        #in the left sub tree. This first element is the smallest element
        
        self.k -= 1
        #Next we check if k ==0, if yes, return the answer
        if self.k == 0:
            self.ans = node.val
            return #Return to the last execution
        
        self.inorder(node.right)# traverse the right subtree of the current node. Vallue of k has changed at this point
        
        
```
